## 9.1 matplotlib API 入门

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook

data = np.arange(10)
data
```

```python
plt.plot(data)
```

```python
# 创建画布 -- figure()
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)   # add_subplot 在画布上绘制多个坐标轴 2x2(4个坐标轴)
ax2 = fig.add_subplot(2, 2, 2)   # 选中当前第 2 个画布
ax3 = fig.add_subplot(2, 2, 3)   # 选中当前第 3 个画布  （所有绘图命令在一个窗口里）
```

```python
plt.plot([1.5, 3, 0.8, 1.6])     # 默认会使用最后一个画布 ax1
```

```python
plt.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)      # ax1 绘制直方图 hist
```

```python
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))    # ax2 绘制散点图 scatter
```

### plt.subplots() 绘制多个坐标轴

```python
# plt.subplots(nrows=2, ncols=2, figsize=(20, 8), dpi=100)
fig, axes = plt.subplots(2, 3)   # subplots() 返回 fig:图对象  ax:坐标轴对象列表
axes
```

```python
# 调整 subplot 周围的间距 -- subplots_adjust()
fig, axes = plt.subplots(2, 2)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
# plt.subplots_adjust(left=None, bottom=None, right=None, tp=None, wspace=0, hspace=0)
plt.subplots_adjust(wspace=0, hspace=0)
```
